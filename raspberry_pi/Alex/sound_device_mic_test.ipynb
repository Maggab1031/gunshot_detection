```python
import sys
import time
import queue

import sounddevice as sd
import soundfile as sf

RECORD_TIME = 5

q = queue.Queue()
rec_start = int(time.time())

sd.default.device = 6
sd.default.samplerate = 48000
sd.default.channels = 1

def data_callback(input_data, frames, time, status):
    if status:
        print(status, file=sys.stderr)
    q.put(input_data.copy())

with sf.SoundFile('SoundDevice Output from Mic.wav', mode='x', samplerate=48000, channels=1) as file:
    with sd.InputStream(callback=data_callback, blocksize=24000, dtype="int16"):
        rec_time = int(time.time()) - rec_start
        
        while rec_time <= RECORD_TIME:
            data = q.get()
            print(data)
            file.write(data)
            print(file.subtype)
            rec_time = int(time.time()) - rec_start
```

```python
import pyaudio
import numpy as np
import soundfile as sf
import librosa
import IPython.display as ipd

pa = pyaudio.PyAudio()
    
stream = pa.open(format = pyaudio.paInt16,
                 rate = 48000,
                 channels = 1,
                 input_device_index = 6,
                 frames_per_buffer = 24000,
                 input = True)
```

```python
sound_data = []
    
with sf.SoundFile('SoundDevice Output from Mic.wav', mode='x', samplerate=48000, channels=1) as file:

    # Loops through the stream and appends audio chunks to the frame array
    for i in range(0, int(48000 / 24000 * 2)):
        sound_buffer = stream.read(24000, exception_on_overflow = False)
        sound_data.append(np.frombuffer(sound_buffer, dtype=np.int16))
    microphone_data = np.concatenate(sound_data)
    file.write(microphone_data)
```
